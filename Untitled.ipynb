```python
import pandas as pd
```

```python
def get_file(col):
    tweets_files = []
    for d in filter(lambda d: 'T' in d, glob('data/*')):
        for f in glob(d+"/*.csv"):
            print(f)
            csv = pd.read_csv(f)
            tweets_files.extend(d+'/'+csv[col])
    return tweets_files
```
